```python
import pandas as pd
import ast
import string
from collections import  Counter
```

```python
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('lmzd_data2.csv', sep=',')
```

```python
df = df.drop(['Unnamed: 0'], axis =1 )
```

```python
punct = [x for x in string.punctuation]
```

```python
df['message'] = df['message'].apply(lambda x: ast.literal_eval(x))
```

```python
for i, text in zip(df.index.values, df['message']):
    df['message'][i] = [t for t in text if not t in punct]
```

```python
all_w = []

for mess in df['message']:
   for word in mess:
      all_w.append(word)
```

```python
# all_uniq_w = set(all_w)
all_uniq_w = Counter(all_w)
qwe = dict(all_uniq_w)
type(qwe)
```

```python
fig = px.scatter(x=all_uniq_w.values(), y=all_uniq_w.keys())
fig.show()
```

```python
word_vals_c = Counter(all_uniq_w.values())

fig = px.scatter(x=word_vals_c.values(), y=word_vals_c.keys())
fig.show()
```

```python
sortedd = dict(sorted(all_uniq_w.items(), key=lambda item: item[1]))
fig = px.scatter(x=sortedd.keys(), y=sortedd.values())
fig.show()
```
